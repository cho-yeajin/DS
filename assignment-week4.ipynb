# 2. LeetCode 232

```python
class MyQueue(object):

    def __init__(self):
        self.stack1=[]
        self.stack2=[]


    def push(self, x):
        self.stack1.append(x)
       

    def pop(self):
        if self.stack2==[] and self.stack1!=[]:
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        if self.stack2 !=[]:
            return self.stack2.pop()
        else:
            return None

    def peek(self):
        if self.stack2==[] and self.stack1!=[]:
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        if self.stack2 !=[]:
            return self.stack2[-1]
        else:
            return None
          
       
    def empty(self):
        return not self.stack1 and not self.stack2
```
